```python
def get_page_num(soup,class_name):
    page_number = 0
    for item in soup.find_all('div', attrs={'class':class_name}):
        for page_num in item.find_all('li', attrs={'class':'ipsPagination_last'}):
            last_page_link = page_num.a["data-page"]
            page_number = int(last_page_link)
    return page_number
```

```python
def get_links(page_number,url):    
    ret_links = []
    counter = 1
    while counter <= 2:
        new_url = url + "?page=" +  str(counter) 
        #print(new_url)
        new_req = urllib.request.Request(new_url, headers={'User-Agent':'Mozilla/5.0'})
        soup = BeautifulSoup(urlopen(new_req).read(),"html.parser") 
        for item in soup.find_all('div', attrs={'class':'ipsBox'}):
            for lines in item.find_all('h4',attrs={'class':'ipsDataItem_title ipsContained_container'}):
                for links in lines.find_all('span',attrs={'class':'ipsType_break ipsContained'}):
                    link = links.a["href"]
                    ret_links.append(link)
        counter+=1
    return ret_links
```

```python
import time

def store_articles(links):
    counter = 356
    name = "mycarforumpost"
    for link in links:
        page_count = 1
        s_req = urllib.request.Request(link, headers={'User-Agent':'Mozilla/5.0'})
        link_soup = BeautifulSoup(urlopen(s_req).read(),"html.parser")
        page_num = get_page_num(link_soup,'ipsSpacer_top ipsClearfix ipsClear')
        to_store = ""
        while page_count <= page_num:
            post_link = link + "?page=" + str(page_count)
            print(post_link)
            s_req = urllib.request.Request(post_link, headers={'User-Agent':'Mozilla/5.0'})
            link_soup = BeautifulSoup(urlopen(s_req).read(),"html.parser")
            for item in link_soup.find_all('div',attrs={'class':'cTopic ipsClear ipsSpacer_top'}):
                for articles in item.find_all('div',attrs={'id':'elPostFeed'}):
                    for article in articles.find_all('article'):
                        to_store += article.find('a')['title'] + article.find('a').text
                        if article.find('span',attrs={'class':'ipsRepBadge ipsRepBadge_positive'}):
                            to_store += article.find('span',attrs={'class':'ipsRepBadge ipsRepBadge_positive'})['title']
                            to_store += article.find('span',attrs={'class':'ipsRepBadge ipsRepBadge_positive'}).text
                        if article.find('li',attrs={'class':'ipsType_light'}):
                            to_store += article.find('li',attrs={'class':'ipsType_light'}).text
                        if article.find('div',attrs={'class':'ipsType_normal ipsType_richText ipsContained'}):
                            to_store += article.find('div',attrs={'class':'ipsType_normal ipsType_richText ipsContained'}).text
            page_count+=1
        f= open("mycarforum/" + name + str(counter) +".txt","w+", encoding="utf-8")
        f.write(to_store)
        f.close()
        counter += 1
        time.sleep(30)
```
